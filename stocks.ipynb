```python
import yfinance as yf
```

```python
microsoft = yf.Ticker("MSFT")
amazon = yf.Ticker("AMZN")
apple = yf.Ticker("AAPL")
nvidia = yf.Ticker("NVDA")
google = yf.Ticker("GOOGL")

```

```python
import pandas as pd
import numpy as np
import matplotlib as mp
```

```python
microsoft = microsoft.history(period="5y")
amazon = amazon.history(period="1y")
apple = apple.history(period="1y")
nvidia = nvidia.history(period="1y")
google = google.history(period="1y")
```

```python
microsoft.plot.line(y="Close",use_index=True)
amazon.plot.line(y="Close",use_index=True)
apple.plot.line(y="Close",use_index=True)
nvidia.plot.line(y="Close",use_index=True)
google.plot.line(y="Close",use_index=True)
```
